```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('mapache.jpg', cv2.IMREAD_GRAYSCALE)
img_convertida = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
plt.imshow(img_convertida)
```

```python
sobel_horizontal = cv2.Sobel(img_convertida, cv2.CV_64F, 1, 0, ksize=5)
plt.imshow(sobel_horizontal)
```
