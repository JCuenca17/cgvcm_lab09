```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread('mapache.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Erosion
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
plt.imshow(erosion)
```

```python
# Erosion Multiple
kernel = np.ones((5,5),np.uint8)
erosionM = cv2.erode(img,kernel,iterations = 4)
plt.imshow(erosionM)
```

```python
# Dilatacion
kernel = np.ones((5,5),np.uint8)
dilatacion = cv2.dilate(img,kernel,iterations = 1)
plt.imshow(dilatacion)
```

```python
# Ruido
white_noise = np.random.randint(low=0, high=2, size=img.shape[:2]) * 255
white_noise_rgb = np.stack([white_noise] * 3, axis=-1)
noise_img = white_noise_rgb + img
plt.imshow(noise_img)
```

```python
# Apertura
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```
