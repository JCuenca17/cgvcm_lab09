```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
img1 = cv2.imread('mapache.jpg')
img2 = cv2.imread('no_copy.png')
img1.shape
```

```python
img2.shape
```

```python
img2 =cv2.resize(img2,(200,200))
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
x_offset = (img1.shape[1] - img2.shape[1]) // 2
y_offset = (img1.shape[0] - img2.shape[0]) // 2
roi = img1[y_offset:y_offset + img2.shape[0], x_offset:x_offset + img2.shape[1]]
plt.imshow(roi)
```

```python
roi[:img2.shape[0], :img2.shape[1]] = img2
plt.imshow(img1)
```
