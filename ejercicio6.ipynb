```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread('mapache.jpg').astype(np.float32) / 255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gamma = 1/4
effected_image = np.power(img, gamma)
plt.imshow(effected_image)
```

```python
gamma = 2
effected_image = np.power(img, gamma)
plt.imshow(effected_image)
```
